## y=X β
对于简单线性回归，向量计法等同于

给定X跟y我们可以使用 NumPy 库解出β值，

```python
from numpy . linalg import inv # 线性几何包，逆矩阵
from numpy import dot, transpose #转置
X = [[1, 6, 2] , [1, 8, 1] , [1, 10, 0] , [1 , 14, 2] , [1, 18, 0]] 
y = [[7] , [9] , [13] , [17.5] , [18]] 
```

```python
print(X)
print(y)
```

```python
print(dot(inv(dot(transpose(X) , X)) , dot(transpose(X) , y))) 
```

NumPy 库也提供了一个最小二乘函数， 它能被用来更简洁地解出参数值

```python
from numpy.linalg import lstsq# 最小二乘
print(lstsq(X,y)[0])
```


**计算给定数组中每行的最大值。**

- `a = np.random.randint(1, 10, [5, 3])`

【知识点：统计相关】
- 如何在二维numpy数组的每一行中找到最大值？

```python
import numpy as np

np.random.seed(100)
a = np.random.randint(1, 10, [5, 3])
print(a)
# [[9 9 4]
#  [8 8 1]
#  [5 3 6]
#  [3 3 3]
#  [2 1 9]]

b = np.amax(a, axis=1)
print(b)
# [9 8 6 3 9]
```

```python
#axis参数的理解
c = np.amax(a,axis=0)
print(c)
```
